## Pergunta de Negócio 2:

### Qual o Total de Vendas Por Data do Pedido?

Demonstre o resultado através de um gráfico de barras.

```python
office = df_dsa.groupby('Data_Pedido')['Valor_Venda']
data = office.count()
data2 = data.sort_index()
grafico = pd.DataFrame(data2)
grafico.plot()
```

## Pergunta de Negócio 3:

### Qual o Total de Vendas por Estado?

Demonstre o resultado através de um gráfico de barras.

```python
office_3 = df_dsa.groupby('Estado')['Valor_Venda']
office_4 = office_3.count()
pd.DataFrame(office_4)
office_4.plot.pie()
```

## Pergunta de Negócio 4:

### Quais São as 10 Cidades com Maior Total de Vendas?

Demonstre o resultado através de um gráfico de barras.

```python
office5 = office_4.sort_values(ascending=False)
print(office5)
print('\nAs 10 cidades com maior total de vendas é: California, New York, Texas, Pennsylvania, Washington, Illinois, Ohio, Florida, Michigan, North Carolina')
```
